```python
# 1D Mesh generieren
ngmesh = NGMesh()
ngmesh.dim = 1

nel = 200
l0 = 2

pnums = []
for i in range(0, nel+1):
    pnums.append (ngmesh.Add(MeshPoint (Pnt(i/nel*l0, 0, 0))))

for i in range(0,nel):
    ngmesh.Add (Element1D([pnums[i],pnums[i+1]], index=1))

ngmesh.Add (Element0D(pnums[0], index=1))
ngmesh.Add (Element0D(pnums[nel], index=2))

mesh = Mesh(ngmesh)
```

```python
# Startwerte:
t0 = 0
dt0 = 2.5e-1
tend = 2
eps0 = 1e-6

# Initialisierung der Parameter
t = Parameter (t0)
dt = Parameter (dt0)
eps = Parameter (eps0)
```

```python
order = 2
V = H1(mesh, order=order)
```

```python
# List fr die gfu's initialisieren
gfu_array = []

for p in range (2,7):
    eps.Set(10**(-p))
    t.Set(t0)

    u = V.TrialFunction()
    v = V.TestFunction()
    gfu = GridFunction(V)
    a = BilinearForm (V, symmetric = False)
    a += Variation((eps/2*grad(u)*grad(u) 
                   +(u**2*(1-u*u)**2)
                   + 0.5/dt*(u-gfu)**2)*dx)
    
    # Anfangswerte setzen
    gfu.Set(-2+4*x)
    
    res = gfu.vec.CreateVector()
    du = gfu.vec.CreateVector()

    # Newton Verfahren:
    for it in range(int((tend-t0)/dt.Get())):
        # Newton Verfahren
        for k in range(15):
            print("Newton iteration", k)    
            # Lsen des linearisierten Problems:
            # F(u_n)
            a.Apply(gfu.vec, res)
        
            # F'(u_n)   
            a.AssembleLinearization(gfu.vec)
        
            # F'(u_n) du = F(u_n)
            du.data = a.mat.Inverse(V.FreeDofs()) * res
        
            # update iteration
            # u_n+1 = u_n - du
            gfu.vec.data -= du
        
            # stopping criteria
            stopcritval = sqrt(abs(InnerProduct(du, res)))
            print("<A u", it, ", A u", it, ">_{-1}A0.S = ", stopcritval)
            print(gfu)
            if stopcritval < 1e-5:
                break
            
    
        # Abbruchkriterium: nach 2 Sekunden (wie in Aufgabenstellung beschrieben)
        t.Set(t.Get() + dt.Get())
        print("abgelaufene Zeit: ", t.Get())
        if t.Get() >= tend and stopcritval < 1e-5:
            break
    
    gfu_array.append(gfu)

```
